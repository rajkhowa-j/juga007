# $L^p$ Spaces

**Definition:** A real function $\phi$ defined on a segment $(a,b)$, where $-\infty\leq a<b\leq\infty$, is called **convex** if the inequality
$$ 
  \phi((1-t)x+ty)\leq (1-t)\phi(x)+t\phi(y)
$$

holds whenever $a<x<b$, $a<y<b$, and $0\leq t\leq 1$.

Graphically, the condition is that if $x<t<y$, then the point $(t,\phi(t))$ should lie below or on the line connecting the points $(x,\phi(x))$ and $(y,\phi(y))$ in the plane. The equation (1) is equivalent to the requirements that
$$
  \frac{\phi(t)-\phi(s)}{t-s}\leq\frac{\phi(u)-\phi(t)}{u-t}
$$

whenever $a<s<t<u<b$.
